```python
# Decision Tree Classification
# Dataset : Iris.csv https://www.kaggle.com/uciml/iris
# We'll try to predict the species of Iris using some features
```

```python
# Import the libraires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.info()
```

```python
# Data preprocessing and splitting into features and classes
df1 = df.iloc[:, [1,3,5]]
x = df1.iloc[:, 0:2].values
y = df['Species'].map({'Iris-setosa': 0, 'Iris-versicolor':1, 'Iris-virginica':2})
```

```python
# Feature scaling on the features
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x = sc_x.fit_transform(x)
```

```python
# Train model on the dataset
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(max_depth = 5)
classifier = classifier.fit(x, y)
```

```python
# Visualising the Test set results
plt.rcParams['figure.figsize'] = [15, 10]
from matplotlib.colors import ListedColormap
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
X_set, y_set = x, y
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.7, cmap = ListedColormap(('red', 'green', 'purple')) ) # was having problems with ListedColormap, don't know why
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'purple'))(i), label = labels[j])
plt.title('Decision Tree')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```

```python
# Errors calculation
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, classifier.predict(x))
err = 0 
for i in range(cm.shape[0]):
    for j in range(cm.shape[0]):
        if( i != j ):
            err += cm[i][j]
print("error: ", err)
print("total accuracy: ", 100*(len(x) - err)/len(x))
```

```python
# As we can see, this model has very high accuracy, which could be improved by increasing the max-depth of tree
# However, we see that it tries to fit every single point in the dataset which could lead to overfitting
```

```python
# Visualize tree
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(classifier, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,
                feature_names = ['Sepal length', 'Petal length'],
                class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                )
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png()) 
```
