```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv('Mall_Customers.csv')
x = df.iloc[:, 3:5].values
```

```python
df.head()
```

```python
# Show dendrogram
plt.rcParams['figure.figsize'] = [15, 10]
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances between clusters')
```

```python
# Taking the optimal split as 5
optimal_split = 5

# Train model
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage = 'ward')
predictions = ac.fit_predict(x)
```

```python
# Visualize
colors = ['red', 'blue', 'green', 'cyan', 'teal']
for i in range(optimal_split):
    plt.scatter(x[predictions == i, 0], x[predictions == i, 1], color = colors[i])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend(['Careful', 'Sensible', 'High spenders', 'Average spenders', 'Careless'])

# Gives same result as with k means
```
